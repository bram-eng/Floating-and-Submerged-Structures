## 2.1 Candidate Concepts

| ID | Name | Description |
|----|------|-------------|
| A  | Pontoon-supported | Continuous pontoons at the water surface support the tunnel tube |
| B  | Tether-moored (SFT) | Submerged floating tunnel anchored by vertical tethers to the seabed |
| C  | Hybrid (SFT + pontoon) | Combination of surface pontoons and submerged tethers |
| D  | Immersed tube | Conventional immersed-tube tunnel lying on the seabed |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

%matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

## 2.2 Multi-Criteria Decision Analysis (MCDA)

### Criteria and Weights

Criteria are defined following PIANC and ISO standards for submerged structures:

| Criterion | Description | Weight |
|-----------|-------------|--------|
| C1 | Structural safety (extreme loads) | 0.25 |
| C2 | Construction feasibility | 0.20 |
| C3 | Capital cost (CAPEX) | 0.20 |
| C4 | Environmental impact | 0.15 |
| C5 | Maintenance & operational cost | 0.10 |
| C6 | Navigation / ship-strike risk | 0.10 |


```python
# -----------------------------------------------------------------------
# Define criteria weights (sum = 1)
# -----------------------------------------------------------------------
criteria = ['Structural safety', 'Construction', 'CAPEX', 'Env. impact', 'O&M cost', 'Nav. risk']
weights = np.array([0.25, 0.20, 0.20, 0.15, 0.10, 0.10])
assert np.isclose(weights.sum(), 1.0), "Weights must sum to 1"

# -----------------------------------------------------------------------
# Score matrix: rows = alternatives, columns = criteria
# Scores on a 1–5 scale (5 = best)
# -----------------------------------------------------------------------
# TODO: Replace placeholder scores with team-assessed values
alternatives = ['A – Pontoon', 'B – SFT tether', 'C – Hybrid', 'D – Immersed tube']
scores = np.array([
    # C1   C2   C3   C4   C5   C6
    [  3,   3,   3,   2,   3,   2 ],   # A – Pontoon
    [  4,   3,   3,   4,   3,   5 ],   # B – SFT tether
    [  4,   2,   2,   3,   3,   4 ],   # C – Hybrid
    [  4,   4,   4,   3,   4,   5 ],   # D – Immersed tube
], dtype=float)

# -----------------------------------------------------------------------
# Weighted scores
# -----------------------------------------------------------------------
weighted = scores * weights
total_scores = weighted.sum(axis=1)

results = pd.DataFrame(weighted, index=alternatives, columns=criteria)
results['Total'] = total_scores
results.sort_values('Total', ascending=False, inplace=True)
print(results.to_string(float_format='{:.3f}'.format))
```

```python
# -----------------------------------------------------------------------
# Spider / radar chart
# -----------------------------------------------------------------------
N = len(criteria)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # close the polygon

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'projection': 'polar'})
colors = cm.tab10(np.linspace(0, 1, len(alternatives)))

for alt, s, c in zip(alternatives, scores, colors):
    vals = s.tolist() + s[:1].tolist()
    ax.plot(angles, vals, 'o-', linewidth=1.5, color=c, label=alt)
    ax.fill(angles, vals, alpha=0.08, color=c)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(criteria, size=9)
ax.set_ylim(0, 5)
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_title('MCDA Spider Chart', y=1.08)
ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.1), fontsize=8)
plt.tight_layout()
plt.show()
```

## 2.3 Life Cycle Analysis (LCA)

A simplified qualitative LCA following ISO 14040/44 framework:

| Phase | Key activities | Main environmental aspects |
|-------|---------------|----------------------------|
| **Materials** | Steel, concrete, HDPE fabrication | Embodied carbon, resource depletion |
| **Construction** | Marine installation, dredging | GHG emissions, seabed disturbance |
| **Operation** | Lighting, ventilation, monitoring | Energy consumption |
| **Maintenance** | Inspection, cathodic protection | Chemical use, vessel emissions |
| **End-of-life** | Decommissioning, recycling | Waste, demolition emissions |


```python
# -----------------------------------------------------------------------
# Illustrative embodied carbon estimate [tonne CO2-eq]
# TODO: Replace with project-specific material take-off
# -----------------------------------------------------------------------
lca_phases = ['Materials', 'Construction', 'Operation (100yr)', 'Maintenance', 'End-of-life']

# Placeholder values [1000 t CO2-eq] — replace with actual LCA data
carbon_est = {
    'A – Pontoon':      [280, 90, 120, 40, -30],
    'B – SFT tether':   [220, 110, 80,  35, -25],
    'D – Immersed tube':[400, 130, 95,  50, -35],
}

x = np.arange(len(lca_phases))
width = 0.25
fig, ax = plt.subplots(figsize=(10, 5))

for i, (name, vals) in enumerate(carbon_est.items()):
    ax.bar(x + i * width, vals, width, label=name, alpha=0.8)

ax.axhline(0, color='k', linewidth=0.8)
ax.set_xticks(x + width)
ax.set_xticklabels(lca_phases, rotation=20, ha='right')
ax.set_ylabel('Carbon [×10³ t CO₂-eq]')
ax.set_title('Indicative Life Cycle Carbon – Concept Comparison')
ax.legend()
ax.grid(True, axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
